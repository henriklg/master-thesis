```python
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from pathlib import Path

import numpy as np
import cv2, os
```

Specify folder path and count number of samples

```python
masked_path = Path('/home/henrik/master_thesis/data/Kvasir-SEG/masked_images/')
num_images = len([name for name in os.listdir(masked_path) if os.path.isfile(os.path.join(masked_path, name))])
```

```python
masked_images = []
name_list = []

# iterate over all image paths
for path in masked_path.iterdir():
    if str(path).endswith(".jpg"):
        filename = path.name
        name_list.append(filename)
        masked_images.append(cv2.imread(str(path)))
```

```python
X = masked_images
y = np.ones(1000)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```
